Lets play a game. How fas t can we get around a board game board? We'll start at position one, roll the die, and move forward that number of spaces. We keep rolling until we have gone all the way around the whole board.

Lets see how we can simulate this game in Python. A foor loop can repeat a task, like rolling the die and advancing the position. But foor loops execute a fixed number of times
Will five rounds be enough to win the game? Ten? Twnty?

How do we know how many times we need to roll the die? We dont. This is where a while loop comes in. A while loop repeatedly executes a chunk of code while a condition is true. For example, here its checking if the current position is less then the total number of squares.  When the condition changes to false it stops.

Basically as long as the current position hasnt reached the end of the board, we keep rolling the die and mocing forward. Once we have made it to the end, the conditin changes and the loop stops repeating.

If you know exactly how many time you need your code to be repeated, use  a for loop. But if you want it repeated until some condition changes, use a while loop.

## Syntax
Eine while-Schleife hat die folgende Struktur:
``` python
while condition:
    statement 1
    statement 2
    statement 3
    ...

another statement 1
another statement 2
...
```
Der Code `statement 1, 2, 3, ...` wird ausgeführt solange die Bedingung `condition` wahr, also `True` ist. Es ist an dieser Stelle wichtig, den Code unterhalb des Schlüsselworts `while` einzurücken, zum Beispiel durch ein Leerzeichen oder ein Tab, da dies Python signalisiert, dass der Code nur ausgeführt werden soll, wenn die Bedingung wahr ist. Der Code `another statement 1, 2, ...` wir unabhänging von dem Wahrheitsgehalt der Bedingung `condition` aus geführt, denn dieser Code gehört nicht zum Schleifenrumpf.


Angenommen wir wollen für ein festes $x$ die natürliche Zahl $n$ zu bestimmen so, dass $y = \frac{x}{2^n}$ kleiner als $1$ ist. Sie können eine while-Schleife schreiben und $x$ durch $2^n$  im Body der Schleife teilen solange der Quotient größer als $1$ ist.
``` python

n = 0
r = x # x / 2^0
while y > 0:
    y = x / ( 2**n)
    n = n + 1
```

## Aufgabe 1 - Maschienengenauigkeit
Computer sind nicht in der Lage Zahlen beliebiger Genauigkeit darzustellen und zu speichern. Daher werden nur eine bestimmte, festgelegte Anzahl an Dezimalstellen gespeichert. Wenn zum Beispiel maximal zwei Dezimalstellen gespeichert werden, dann sind die Zahlen $0.001$ und $0.008$ beide "gleich" Null.


Was ist der größte Wert für $\varepsilon$, sodass Python nicht zwischen $1$ und $1 + \varepsilon$ unterscheiden kann?

:::{admonition} Aufgabe 1
Modifizieren Sie den nachfolgenden Code so, dass $\varepsilon$ so lange halbiert wird bis für Python $1 + \varepsilon$ gleich $1$ ist.
:::

````python
x = 1
epsilon = x
y = x + epsilon


epsilon = epsilon / 2
y = x + epsilon

print(str(x) + " + " + str(epsilon) + " is the same as " + str(x))
````

:::{admonition} Hinweis
:class: note dropdown

Nutzen Sie eine while-Schleife. Der folgende Code wird ausgeführt bis $k > 5$.
``` python

```
Iteriren Sie so lange bis $y$ (also $ 1 + \varepsilon$) ungleich $x$ (also $1$) ist. Die Python-Syntax für den logischen Operator $\neq$ ist `!=`.
:::

:::{admonition} Lösung
:class: tip dropdown

``` python

    epsilon = epsilon / 2

```
:::

Beachten Sie, dass $\varepsilon$ in der Größenordnung von $10^{-16}$ liegt.


## Aufgabe 2 - Best-of-Five

Um bei einem Spiel mit der Gewinnregel „best out of five“ müssen Sie nur drei Runden gewinnen. Ein Sieger kann in drei, vier oder fünf Runden ermittelt werden, und Sie wissen vorab nicht wie viele Runden Sie brauchen werden.
Der nachfolgende Code modelliert eine Runde, indem er einen Sieger bestimmt ($1$ für Team A und $2$ für Team B). Dann werden die bisherigen Gesamtsiege für jedes Team in `teamA` und `teamB` gespeichert. 

:::{admonition} Aufgabe 2.1
Modifizieren Sie den Code so, dass dieser so lange ausgeführt wird, bis entweder Team A oder Team B drei Spiele gewonnen hat. 
:::


````python
import numpy as np

k = 0
teamA = 0
teamB = 0
outcome = np.zeros(5)

outcome[k] = np.random.randint(1, 3)
teamA = np.sum(outcome == 1)
teamB = np.sum(outcome == 2)  
k = k + 1
````

:::{admonition} Hinweis
:class: note dropdown

Überprüfen Sie ob sowohl Team A als auch Team B bisher weniger als dreimal gewonnen hat. Ein Möglichkeit ist das Maximum von `teamA` und `teamB` zu betrachten (`np.max([teamA, teamB]) < 3`).
:::

:::{admonition} Lösung

import numpy as np
